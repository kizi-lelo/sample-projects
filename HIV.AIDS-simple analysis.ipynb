#  visualization for Tanzania

visualizing data for Tanzania requires to fetch data it in column (entity)

what is the number of new cases and rate of death by age?

```python
df_new_cases_tz=df_new_cases[df_new_cases['Entity']=='Tanzania']
```

```python
df_new_cases_tz
```

```python
df_new_cases_tz.plot.bar(x='Year', y='Incidence - HIV/AIDS - Sex: Both - Age: All Ages (Number)')
```

```python
df_death_rate_tz=df_death_rate[df_death_rate['Entity']=='Tanzania']
```

```python
df_death_rate_tz.set_index('Year').plot(figsize=(15,10))
```

#  visualization worldwide

```python
#top 5 countries with much number of new cases

df_new_cases.groupby('Entity')['Incidence - HIV/AIDS - Sex: Both - Age: All Ages (Number)'].sum().sort_values(ascending=False).head(5).plot.bar(title=('country with more new cases'),figsize=(15,10))
```

```python
#top 5 countries with less number of new cases
df_new_cases.groupby('Entity')['Incidence - HIV/AIDS - Sex: Both - Age: All Ages (Number)'].sum().sort_values(ascending=True).head(5).plot.bar(title=('country with new less cases'),figsize=(15,10))
```
